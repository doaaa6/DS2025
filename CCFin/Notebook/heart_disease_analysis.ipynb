```python
from ucimlrepo import fetch_ucirepo

# fetch dataset
heart_disease = fetch_ucirepo(id=45)

# data (as pandas dataframes)
X = heart_disease.data.features
y = heart_disease.data.targets

# metadata
print(heart_disease.metadata)

# variable information
print(heart_disease.variables)
```

```python
# ==============================
# ANALYSE DU DATASET HEART DISEASE (UCI)
# ==============================

# 1️⃣ Importation des librairies
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from ucimlrepo import fetch_ucirepo

# 2️⃣ Chargement du dataset
heart_disease = fetch_ucirepo(id=45)

# Données principales
X = heart_disease.data.features
y = heart_disease.data.targets

# Fusion pour analyse complète
data = pd.concat([X, y], axis=1)

# 3️⃣ Aperçu des métadonnées
print("=== MÉTADONNÉES ===")
print(heart_disease.metadata)
print("\n=== INFORMATIONS SUR LES VARIABLES ===")
print(heart_disease.variables)

# 4️⃣ Exploration des données
print("\nAperçu du jeu de données :")
print(data.head())

print("\nRésumé statistique :")
print(data.describe())

# 5️⃣ Vérification des valeurs manquantes
plt.figure(figsize=(8, 5))
sns.heatmap(data.isnull(), cbar=False, cmap="Reds")
plt.title("Valeurs manquantes dans le dataset")
plt.show()

# 6️⃣ Distribution de la variable cible (présence de maladie cardiaque)
plt.figure(figsize=(6, 4))
sns.countplot(data=y, x=y.columns[0], palette="coolwarm")
plt.title("Distribution de la présence de maladie cardiaque")
plt.xlabel("0 = Pas de maladie | 1 = Maladie présente")
plt.ylabel("Nombre de patients")
plt.show()

# 7️⃣ Distribution de l'âge
plt.figure(figsize=(7, 4))
sns.histplot(data['age'], bins=20, kde=True, color='steelblue')
plt.title("Distribution de l'âge des patients")
plt.xlabel("Âge")
plt.ylabel("Fréquence")
plt.show()

# 8️⃣ Corrélation entre variables
plt.figure(figsize=(12, 10))
sns.heatmap(data.corr(numeric_only=True), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Matrice de corrélation des variables cliniques")
plt.show()

# 9️⃣ Relations entre certaines variables et la maladie cardiaque
plt.figure(figsize=(12, 5))
sns.boxplot(data=data, x=y.columns[0], y="age", palette="Set2")
plt.title("Âge selon la présence de maladie cardiaque")
plt.xlabel("Présence de maladie (0 = Non, 1 = Oui)")
plt.ylabel("Âge")
plt.show()

plt.figure(figsize=(12, 5))
sns.violinplot(data=data, x=y.columns[0], y="chol", palette="muted")
plt.title("Niveau de cholestérol selon la maladie cardiaque")
plt.xlabel("Présence de maladie (0 = Non, 1 = Oui)")
plt.ylabel("Cholestérol (mg/dl)")
plt.show()

# 10️⃣ Pairplot pour visualiser des relations multiples
sns.pairplot(data[['age', 'trestbps', 'chol', 'thalach', y.columns[0]]], hue=y.columns[0], palette="husl")
plt.suptitle("Relations entre variables clés et la maladie cardiaque", y=1.02)
plt.show()

print("\n✅ Analyse graphique terminée.")
```
